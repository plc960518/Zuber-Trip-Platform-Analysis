```python
# Importar librerías
from scipy import stats as st
from scipy.stats import ttest_ind
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import levene
```

```python
# Datasets
df_company_trips = pd.read_csv('data\project_sql_company_name.csv')
df_location_trips = pd.read_csv('data\project_sql_trips_location.csv')
```

```python
print("DF Company Trips Info")
df_company_trips.info()
print(df_company_trips.head())

print("DF Location Trips Info")
df_location_trips.info()
print(df_location_trips.head())
```

# Top 10 barrios en finalización de recorridos

```python
top_10_barrios = df_location_trips.sort_values(by= 'average_trips', ascending=False)
print(top_10_barrios.head(10))
```

```python
# Gráfico de barras para los top 10 barrios
plt.figure(figsize=(12, 6))
plt.bar(top_10_barrios.head(10)['dropoff_location_name'],top_10_barrios.head(10)['average_trips'], color='lightcoral')
plt.title('Top 10 Barrios por finalización de recorridos', fontsize=16)
plt.xlabel('Barrios')
plt.ylabel('Promedio de viajes')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Top 10 empresas de taxis por cantidad de viajes

```python
top_10_empresas = df_company_trips.sort_values(by= 'trips_amount', ascending=False)
print(top_10_empresas.head(10))
```

```python
# Gráfico de barras para las top 10 empresas de taxis
plt.figure(figsize=(12, 6))
plt.bar(top_10_empresas.head(10)['company_name'],top_10_empresas.head(10)['trips_amount'], color='lightblue')
plt.title('Top 10 Empresas de taxis por cantidad de viajes', fontsize=16)
plt.xlabel('Empresas de taxis')
plt.ylabel('Cantidad de viajes')
plt.grid(True)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

# Distribución de viajes por empresa de taxis

```python
# Distribución de viajes por empresa de taxis
plt.figure(figsize=(10,6))
plt.hist(df_company_trips['trips_amount'], bins=15, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('Distribución de viajes por empresa', fontsize=16)
plt.xlabel('Cantidad de viajes')
plt.ylabel('Número de empresas')
plt.grid(True, alpha=0.3)
plt.show()
```

# Conclusiones análisis exploratorio:
En el gráfico de los 10 barrios con mayor número de recorridos finalizados, se aprecia una clara concentración de actividad en cuatro zonas principales: Loop, River North, Streeterville y West Loop. Estos barrios presentan promedios por encima de los 5,000 viajes, muy por encima del resto, que se mantiene en un rango aproximado de 1,500 a 2,500 viajes. Esta diferencia puede estar relacionada con la alta concentración de oficinas, servicios y puntos de interés que atraen mayor flujo de pasajeros.

Respecto a las empresas de taxis, el comportamiento es más extremo. Flash Cab destaca de manera notable con casi 20,000 viajes, superando de forma considerable incluso a las compañías incluidas en el top 10, cuyo rango se sitúa entre casi 6,000 y 11,500 viajes. Esto también se refleja en el histograma, donde la distribución principal se observa entre 2,500 y 11,000 viajes, aunque la mayoría de las empresas se ubican por debajo de 1,250 viajes, mostrando una participación mucho menor en el total de recorridos.

En conjunto, estos resultados permiten identificar tanto la alta concentración geográfica de la demanda como la desigual participación entre las empresas, lo cual constituye un punto de partida relevante para profundizar en análisis posteriores.
