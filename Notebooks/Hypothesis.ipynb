```python
# Load DF for weather conditions
df_hypothesis = pd.read_csv('data\project_sql_result_weather.csv')
df_hypothesis.info()
```

```python
print(df_hypothesis.sample(5))
print("\nCondiciones climáticas únicas:")
print(df_hypothesis['weather_conditions'].unique())
```

```python
# Converción de datos start_ts a datetime y extracción del día sábado
df_hypothesis['start_ts'] = pd.to_datetime(df_hypothesis['start_ts'])
df_hypothesis['dow'] = df_hypothesis['start_ts'].dt.day_name()

# Verificar los días de la semana disponibles
print("Días de la semana únicos:")
print(df_hypothesis['dow'].unique())
```

```python
# Categorizar condiciones de clima en 'malos' y 'buenos'
rainy_sat = df_hypothesis[(df_hypothesis['dow'] == 'Saturday') & (df_hypothesis['weather_conditions'] == 'Bad')]['duration_seconds']
other_days = df_hypothesis[~((df_hypothesis['dow'] == 'Saturday') & (df_hypothesis['weather_conditions'] == 'Bad'))]['duration_seconds']

print(f"Sábados lluviosos: {len(rainy_sat)} viajes")
print(f"Otros días/condiciones: {len(other_days)} viajes")
```

## Estadísticas descriptivas de las categorías

```python
print("Estadísiticas: Sábados lluviosos")
print(f"Duración promedio: {rainy_sat.mean():.2f} segundos")
print(f"Desviación estándar: {rainy_sat.std():.2f} segundos")
print(f"Duración mínima: {rainy_sat.min():.2f} segundos")
print(f"Duración máxima: {rainy_sat.max():.2f} segundos")

print("\nEstadísticas: Otros días/condiciones")
print(f"Duración promedio: {other_days.mean():.2f} segundos")
print(f"Desviación estándar: {other_days.std():.2f} segundos")
print(f"Duración mínima: {other_days.min():.2f} segundos")
print(f"Duración máxima: {other_days.max():.2f} segundos")
```

## Boxplot para comparación visual

```python
plt.figure(figsize=(10, 6))
plt.boxplot([rainy_sat, other_days], 
            labels=['Sábados lluviosos', 'Otros días'])
plt.title('Comparación de duración de viajes')
plt.ylabel('Duración (segundos)')
plt.show()
```

```python
# Prueba de igualdad de varianzas
stat_levene, p_value_levene = levene(rainy_sat, other_days)
print(f"Estadístico de Levene: {stat_levene:.4f}")
print(f"P-valor: {p_value_levene:.4f}")

if p_value_levene < 0.05:
    print("Las varianzas son significativamente diferentes")
    equal_var = False
else:
    print("Las varianzas son iguales")
    equal_var = True

print()
stat, p_value = ttest_ind(rainy_sat, other_days, equal_var=True)
alpha = 0.05
print(f"Estadístico t: {stat:.4f}")
print(f"Valor p: {p_value:.4f}")
print(f"Nivel de significacia (alpha): {alpha}")
```

## Prueba de hipótesis

```python
# Prueba las hipótesis
if p_value < alpha:
    print(f"Conclusión: Rechazamos la hipótesis nula (p-value = {p_value:.4f} < {alpha})")
    print("Existe evidencia estadística significativa de que la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos.")
else:
    print(f"\nConclusión: No rechazamos la hipótesis nula (p-value = {p_value:.4f} >= {alpha})")
    print("No hay evidencia estadística suficiente de que la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos.")
```
